```python
def SigGenes(InDic):
    newList=[]
    for key in InDic:
        for k in key:
            if k in newList:
                pass
            else:
                newList.append(k)
    return newList
T_sig=SigGenes(Throt_STD_all)
S_sig=SigGenes(Skcos_STD_all)
```

```python
def Kegg_parser(k_args, kegg_genes):
    if __name__=="__main__":

#         k_args=sys.argv[1]
#         kegg_genes=sys.argv[2]	
#         CountFile=sys.argv[3]
#         outFile=sys.argv[4]	
#         f=open(outFile, 'w')

    # """	
    #  PARSE THE ko00002.keg file to get dictionary of dictionaries for each of the levels. 
    # """
        A_2_B_Dict={}
        B_2_C_Dict={}
        C_2_D_Dict={}
        D_2_E_Dict={}

        for line in open(k_args):
            line=re.sub('<[^<]+?>', '', line)
            if line.startswith("A"):
                A=line.strip("A").strip()
            elif line.startswith("B"):
                line=line.strip().strip("B").strip()
                if line!='':
                    b=line
                    if A in A_2_B_Dict:
                        A_2_B_Dict[A]+=[b]
                    else:
                        A_2_B_Dict[A]=[b]
            elif line.startswith("C"):
                line=line.strip().strip("C").strip()
                if line!='':
                    c=line
                    if b in B_2_C_Dict:
                        B_2_C_Dict[b]+=[c]
                    else:
                        B_2_C_Dict[b]=[c]
            elif line.startswith("D"):
                line=line.strip().strip("D").strip()
                if line!='':
                    d=line.strip().split()[0]
                    if c in C_2_D_Dict:
                        C_2_D_Dict[c]+=[d]
                    else:
                        C_2_D_Dict[c]=[d]

            elif line.startswith("E"):
                line=line.strip().strip("E").strip()
                if line!='':
                    e=line.strip().split()[0]
                    if d in D_2_E_Dict:
                        D_2_E_Dict[d]+=[e]
                    else:
                        D_2_E_Dict[d]=[e]
    # """	
    #  PARSE THE KAAS file (alignment to the KEGG directory of transcripts); link the K#### to the gene id from our data
    # """
        KASS_Gene_Dict={}
        KASS_Gene_Dict['Unknown']=[]
        for line in open(kegg_genes):
            line=line.strip().split()
            if len(line)==1:
                KASS_Gene_Dict['Unknown']+=line
            else:
                KASS=line[1]
                gene=line[0]
                if KASS in KASS_Gene_Dict:
                    KASS_Gene_Dict[KASS]+=[gene]
                else: 
                    KASS_Gene_Dict[KASS]=[gene]
        Org_gene_Dict={}
        for line in open(kegg_genes):
            line=line.strip().split()
            if len(line)==1:
                pass
            else: 
                KASS=line[1]
                gene=line[0]
                Org_gene_Dict[gene]=KASS
                
    # """	
    #  Get the number of gene represntatives in each of the KOs
    # """
        NumberGenesIn_KASS={}
        for key in KASS_Gene_Dict:
            NumberGenesIn_KASS[key]=len(KASS_Gene_Dict[key])
    # 	print NumberGenesIn_KASS[key]
        D_2_E_Num={}	
        c=0
        for j in D_2_E_Dict:
            for gID in D_2_E_Dict[j]:
                if gID in NumberGenesIn_KASS:
                    c=NumberGenesIn_KASS[gID]
                else:
                    c=0
                if j in D_2_E_Num:
                    d=D_2_E_Num[j]
                    cd = c+d 
                    D_2_E_Num[j]=cd
                else:
                    D_2_E_Num[j]=c
        C_2_D_Num={}
        for j in C_2_D_Dict:
            for gID in C_2_D_Dict[j]:
                if gID in D_2_E_Num:
                    c=D_2_E_Num[gID]
                else:
                    c=0
                if j in C_2_D_Num:
                    d=C_2_D_Num[j]
                    cd = c + d
                    C_2_D_Num[j]=cd
                else:
                    C_2_D_Num[j]=c
        for key in C_2_D_Num:
            print key, C_2_D_Num[key]
    rr=[Org_gene_Dict, KASS_Gene_Dict, D_2_E_Dict, C_2_D_Dict, B_2_C_Dict, C_2_D_Num]
    return rr
```

```python
Skcos_KEGG_out=Kegg_parser('ko00002.keg', 'Skcos_KAASOutput.txt')
Throt_KEGG_out=Kegg_parser('ko00002.keg', 'Throt_KAASOutput.txt')
Throt_KEGG_out[4]
```

```python
def countKeggGenes(KEGG_out, SigGenes):
    #Count the total number of genes that are in each kegg module 
        #KEGG_out =  [Org_Gene_Dict, KASS_Gene_Dict, D_2_E_Dict, C_2_D_Dict, B_2_C_Dict]
    SigKegg=[]
    for x in SigGenes:
        if x in KEGG_out[0].keys():
            SigKegg.append(KEGG_out[0][x])
    Count_Hash={}
    for x in KEGG_out[3].keys():
        S_genes=set()
        for y in KEGG_out[3][x]:
            for z in KEGG_out[2][y]:
                if z in KEGG_out[1].keys():
                    for q in KEGG_out[1][z]:
                        if q in SigGenes: 
                            S_genes.add(q)
        c=len(S_genes)
        Count_Hash[x]=[S_genes,c]
    return Count_Hash

def higher_sum(SigHash,KEGGHash):
    outHash={}
    for key in KEGGHash:
        for item in KEGGHash[key]:
            if key in outHash.keys():
                outHash[key]+=SigHash[item][1]
            else:
                outHash[key]=SigHash[item][1]
    return outHash
```

```python
T_Sig_Hash=countKeggGenes(Throt_KEGG_out, T_sig)
S_Sig_Hash=countKeggGenes(Skcos_KEGG_out, S_sig)
# Throt_KEGG_out[4]
T_Sig_Hash_High=higher_sum(T_Sig_Hash,Throt_KEGG_out[4])#Get highest kegg module counts for sig genes
S_Sig_Hash_High=higher_sum(S_Sig_Hash,Skcos_KEGG_out[4])
```
