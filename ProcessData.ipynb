##Identification of Stable and Resource Responsive Genes and the calculation of SGNC and STD

Pipeline to identify Resource Responsive (RR) gene set, identify stable genes, calculate stable gene normalized counts (SGNC), and ultimately calcualte STDN and STDP for each RR gene.  

A question was raised about our choice of fc=1.25/post-p<0.1 cutoff for the selection of Stable Genes(SG) in the calculation of the SGNC. Here I stream line the entire process of NC to SGNC to STD calculation and within this vary the number of SGs used to normalize the SGNC for each organism. We do this by expanding the gene set used to calculate the gene set by varrying the posterior probabiliby cutoff considered from 0.1 to 0.5.

### Functions
A series of functions that are called throughout this notebook that are of general utility. 

```python
#Functions to calculate STD

def import_Table(tabFile):
    #import panda type data table from tab delim
    pdTab=pd.read_table(tabFile,index_col=0)
    return pdTab

def calculateTPM(pdTable):
    #calculate tpm for pandas data table given raw counts
    pdTable=pdTable.astype(float)
    sums=pdTable.sum()
    out=(pdTable/sums)*1e6
    return out

def getASC(postProbDir, suffix):
    #Loop through a directory and import the ASC details for all files in directory
    pwd = os.getcwd()
    os.chdir(postProbDir)
    Hash={}
    files=glob.glob('*'+suffix)
    for f in files:
        #Create a hash of the pdTables that are callable by filename without .txt
        pdTab=pd.read_table(f, index_col=0)
        Hash[f[:-4]]=pdTab 
    os.chdir(pwd)
    return Hash

def cutTable(padTab, cutoff):
    #Input a panda table and it will cut the table taking only values greater than the cutoff
    ind=padTab>cutoff
    t=ind.sum(axis=1)>=1
    Inx=ind[t].index
    pdOut=padTab.loc[Inx]
    
    return pdOut

def Get_Stable(Hash,cutoff):
    #Gets stable genes by looping through incubations of interest and getting the stable genes 
    c=cutoff
    StableGeneSet=set()
    for i in Incubations:
        H=Hash[i+'_125'] #get working pd table
        Hcut=H[cols]<=c
        s=Hcut.sum(axis=1)
        SGene=H[s==2] #stable genes
        SList=SGene.index
        SSet=set(SList)
        if StableGeneSet==set():
            StableGeneSet=SSet
        else:
            H=StableGeneSet.intersection(SSet)
            StableGeneSet=H
    return StableGeneSet

def Get_Variable(InHash, cutoff):
    # Within a Hash containing all of the ASC files and a cutoff value get a new hash containing 
    #a list of all the up and down regulated genes for and P
    c=cutoff
    OutHash={}
    for i,iName in zip(NandP,N):
        H=InHash[i+'_2'] #get working pd table
        for j,jName in zip(cols,U):
            Hcut=H[j]>=c
            VGene=H[Hcut] #variable genes
            VList=VGene.index
            OutHash[iName+'_'+jName]=VList
#         SSet=set(SList)
#         if StableGeneSet==set():
#             StableGeneSet=SSet
#         else:
#             H=StableGeneSet.intersection(SSet)
#             StableGeneSet=H
    return OutHash

def CalcSGNC(padTPM,SGeneList):
    #Input the TPM pd table and a list of stable genes; returns a pd table of SGNC
    S_mean=padTPM.loc[SGeneList].mean() #calculate the mean
#     S_gmean=stats.mstats.gmean(padTPM.loc[SGeneList])
    padSGNC=padTPM/S_mean
#     print S_mean
#     print S_gmean
#     plt.plot(S_mean, S_gmean, '.')
#     plt.show()
    return padSGNC

def CalcSTD(pdSGNC, VGenes):
    #Given a panda dictionary with SGNC values and a hash of lists with the 
    #variable genes will calculate the STDN and STDP for each gene and create a new master panda dataframe
    insitu=['S1','S2','S3','S4','S5']
    S=sorted(insitu*2)
    T=['STDN', 'STDP']*5
    columns=[S,T]
    Hash={}
    for L in VGenes: 
        VG=VGenes[L] #One set of significant genes
        WG=pdSGNC.loc[VG] #Working SGNC pandas dictionary with the significant genes from above
        df=pd.DataFrame(index=VG, columns=columns)
        pNSGNC=WG['A']
        mNSGNC=WG['B']
        pPSGNC=WG['C']
        mPSGNC=WG['D']
        for i in insitu:
            In=WG[i]
            stdN=(In-pNSGNC)/(mNSGNC-pNSGNC)
            stdP=(In-pPSGNC)/(mPSGNC-pPSGNC)
            df.loc[:,(i,'STDN')]=stdN
            df.loc[:,(i,'STDP')]=stdP
        Hash[L]=df
    return Hash

def quadrantCount(STD,cutoffN,cutoffP):
    #input panda dict of STD counts and cutoff value
    insitu=['S1','S2','S3','S4','S5']
    cN=cutoffN
    cP=cutoffP
    Nquad=[0.0]*5
    Pquad=[0.0]*5
    c=0
    for i in insitu:
        for ind in STD.index:
            N=STD.loc[ind,(i,'STDN')]
            P=STD.loc[ind,(i,'STDP')]
            if (N > cN) and (P < cP):
                Nquad[c]+=1
            elif P > cP and N < cN:
                Pquad[c]+=1
        c+=1
    Nquad= [x/len(STD.index) for x in Nquad]   
    Pquad= [x/len(STD.index) for x in Pquad]   
    dfN=pd.DataFrame(index=['NQuad'], columns=insitu)
    dfP=pd.DataFrame(index=['PQuad'], columns=insitu)
    dfN.loc['NQuad']=Nquad
    dfP.loc['PQuad']=Pquad
    
    return dfN,dfP
    
def iterateQuadrant(PDSTD, high, low,step):
    #given high and low bounds; will step over the range and calculate the mean and standard deviation of the qaudran
    R = np.linspace(low,high,step)
    c=0
    for i in R:
        for j in R:
            c+=1
            (dfN,dfP)=quadrantCount(PDSTD,i,j)
            if c == 1:
                catdfN=dfN.copy()
                catdfP=dfP.copy()
            else:
                catdfN=pd.concat((catdfN,dfN))
                catdfP=pd.concat((catdfP,dfP))
    dfOut=pd.DataFrame(columns=dfN.columns,index=['Nmean','Pmean','Nstd','Pstd'])
    dfOut.loc['Nmean']=catdfN.mean()
    dfOut.loc['Pmean']=catdfP.mean()
    dfOut.loc['Nstd']=catdfN.std()
    dfOut.loc['Pstd']=catdfP.std()
    return dfOut
```
